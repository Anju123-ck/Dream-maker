# Importing basic Libraries

```python
#Basic Operations
import pandas as pd
import numpy as np

#For Visualizatins
import seaborn as sns
import matplotlib.pyplot as plt

#To ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading the dataset into the environment
data=pd.read_csv('dataset_players.csv')
```

```python
#Checking the first 5 columns of the dataset
data.head(5)
```

```python
data.tail()
```

# Descriptive Analysis

```python
data.describe()
```

```python
data.shape
```

There are 159541 rows and 89 columns in the dataset.

```python
data.info()
```

There are float,int,object type datatypes in the data.Also there are null values in the dataset.

```python
data.isna().sum()[data.isna().sum()>0]
```

```python
#Filling the null value columns

cols=['IntGoals','U21Caps','Height','Communication','Aggression']

for i in cols:
    data[i].fillna(data[i].mean(),inplace=True)
```

```python
data['PositionsDesc'].fillna( method ='ffill', inplace = True)
```

```python
data['Born'].fillna('not known',inplace=True)
```

```python
data.isna().sum()[data.isna().sum()>0]
```

Now there is no null values in the dataset.

# Exploratory Data Analysis

```python
#Distribution of age of players
x=data['Age']
plt.figure(figsize=(10,8))
ax=sns.countplot(x,color='#00ffff')
ax.set_xlabel(xlabel='Age of Players',fontsize=16)
ax.set_title(label='Distribution of Age of Players',fontsize=20)
plt.show()
```

from the visualization we can see that most of the players are in the age group of 16-27 years

```python
plt.figure(figsize = (20, 8))
plt.style.use('fivethirtyeight')
ax = sns.countplot('PositionsDesc', data = data, palette = 'bone')
ax.set_xlabel(xlabel = 'Different Positions in Football', fontsize = 16)
ax.set_ylabel(ylabel = 'Count of Players', fontsize = 16)
ax.set_title(label = 'Comparison of Positions and Players', fontsize = 20)
plt.show()
```

```python
data['Height']=data['Height'].astype(int)
```

```python
#count of players on the basis of heights
plt.figure(figsize = (25,15))
ax = sns.countplot(x = 'Height', data = data, palette = 'dark')
ax.set_title(label = 'Count of players on Basis of Height', fontsize = 20)
ax.set_xlabel(xlabel = 'Height of the player', fontsize = 16)
ax.set_ylabel(ylabel = 'Count', fontsize = 16)
plt.show()
```

```python
#weight of players
from matplotlib import style

plt.figure(figsize = (20, 5))
plt.style.available
sns.distplot(data['Weight'], color = 'pink')
plt.title('Different Weights of the Players ', fontsize = 20)
plt.xlabel('Weights associated with the players', fontsize = 16)
plt.ylabel('count of Players', fontsize = 16)
plt.show()
```

```python
plt.figure(figsize = (10, 8))
ax = sns.countplot(x = 'Positioning', data = data, palette = 'pastel')
ax.set_title(label = 'Count of players on Basis of their positioning', fontsize = 20)
ax.set_xlabel(xlabel = 'Number of positioning ', fontsize = 16)
ax.set_ylabel(ylabel = 'Count', fontsize = 16)
plt.show()
```

```python
#violin plot
plt.rcParams['figure.figsize'] = (20, 7)
plt.style.use('seaborn-dark-palette')

sns.boxenplot(data['Decisions'], data['Leadership'], hue = data['Age'], palette = 'Greys')
plt.title('Comparison of Decisions and Leadership with respect to Age', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (20, 7)
plt.style.use('seaborn-dark-palette')

sns.boxenplot(data['Stamina'], data['Consistency'], hue = data['ImportantMatches'], palette = 'Greys')
plt.title('Comparison of stamina and consistency with respect to important matches', fontsize = 20)
plt.show()
```
