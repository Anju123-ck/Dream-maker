# males and females participated in the test

```python
sns.set(font_scale=1.5,)
df['gender'].value_counts().plot(kind='bar',color='y',figsize=(7,6),rot=0)
plt.xlabel('gender',labelpad=10)
plt.ylabel('number of participants',labelpad=10,)
plt.title('number of female and male participated in the exam')
```

# students' parental level of education

```python
df['parental level of education'].value_counts().plot(kind='bar',color='g')
```

```python
df['parental level of education'].value_counts()
```

# average for math,reading,writing based on gender

```python
df.groupby('gender')[['math score','writing score','reading score']].mean()
```

```python
female participants get more marks than male in writing and reading
```

# average for math,reading,writing based on test preparation course

```python
df.groupby('test preparation course')[['math score','writing score','reading score']].mean()
```

```python
participants who completed the test preparation course get good marks than who do not done it
```

# scoring variation for math, reading and writing based on gender

```python
df.groupby('gender')[['math score','writing score','reading score']].std()
```

# scoring variation for math,reading and writing based on test preparation course

```python
df.groupby('test preparation course')[['math score','writing score','reading score']].std()
```
