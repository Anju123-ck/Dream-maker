```python
Q1=np.percentile(data['Age'],25,interpolation='midpoint')
Q3=np.percentile(data['Age'],75,interpolation='midpoint')
```

```python
IQR=Q3-Q1
IQR
```

```python
low_lim=Q1-1.5*IQR
upp_lim=Q3+1.5*IQR
```

```python
print(low_lim)
print(upp_lim)
```

```python
outlier=[]
for x in data['Age']:
    if((x>upp_lim) or (x<low_lim)):
        outlier.append(x)
```
