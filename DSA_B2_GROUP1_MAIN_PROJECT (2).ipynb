```python
#Null value detection and filling
data.isna().sum()[data.isna().sum()>0]
```

There are null values in Age,Height,Positionsdesc and bmi columns.

```python
freqgraph=data.select_dtypes(include=['float'])
freqgraph.hist(figsize=(10,8))
plt.show()
```

Age and BMI shows a skewed distribution.Hence it's null values can be filled with median values and Height with mean values,since it shows a normal distribution.

```python
data['Age']=data['Age'].fillna(data['Age'].median())
data['Bmi']=data['Bmi'].fillna(data['Bmi'].median())
data['Height']=data['Height'].fillna(data['Height'].mean())
data['PositionsDesc'] = data['PositionsDesc'].fillna(method = 'ffill')
```

```python
Q1=np.percentile(data['Age'],25,interpolation='midpoint')
Q3=np.percentile(data['Age'],75,interpolation='midpoint')
IQR=Q3-Q1
low_lim=Q1-1.5*IQR
upp_lim=Q3+1.5*IQR

data['Age']=np.where(data['Age']>upp_lim,upp_lim,np.where(data['Age']<low_lim,low_lim,data['Age']))
        
```

```python
#Encoding and feature reduction
#Since we have 84 columns one hot encoding will increase the number of columns to more than 300,which will effect our modelling.
#So we will limit our one hot encoding to the 50 most frequent labels of the variable.
#That is making one binary variable for each of the most 50 variables only.
data.PositionsDesc.value_counts().sort_values(ascending=False).head(50)
```

```python
top_50=[x for x in data.PositionsDesc.value_counts().sort_values(ascending=False).head(50).index]
top_50
```

```python
for label in top_50:
    data[label]=np.where(data['PositionsDesc']==label,1,0)
    
data[['PositionsDesc']+top_50].head()
```

```python
def one_hot_top_x(data,variable,top_x_labels):
    for label in top_x_labels:
        data[variable+'_'+label]=np.where(data[variable]==label,1,0)
        
```

```python
from sklearn import preprocessing

x=data.values
scaler=preprocessing.MinMaxScaler()
x_scaled=scaler.fit_transform(x)
x_norm=pd.DataFrame(x_scaled)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
reduced=pd.DataFrame(pca.fit_transform(x_norm))
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5)
kmeans=kmeans.fit(reduced)
labels=kmeans.predict(reduced)
centroid=kmeans.cluster_centers_
clusters=kmeans.labels_.tolist()
```

```python
reduced['cluster']=clusters
reduced['Name']=Name
reduced.columns=['x','y','cluster','Name']
reduced.head()
```

```python
sns.set(style='white')
ax=sns.scatterplot(x='x',y='y',hue='cluster',data=reduced,legend=False,
              size=15)
texts=[]
for x,y,s in zip(reduced.x,reduced.y,reduced.Name):
    texts.append(plt.text(x,y,s))
    
ax.set(ylim=(-2,2))
plt.tick_params(labelsize=15)
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.show()
```
